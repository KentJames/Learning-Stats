# Standard Deviation

The standard deviation is a way of measuring the spread of data about a central value. It's a defined feature of almost (but not all) probability distributions. For a linear (not circular, more on that later) distribution, it's definition is the square root of the variance.

$$ \sigma = \sqrt{\frac{\Sigma^N_{i=0} \big(x_i - \bar{x}\big)}{N-1}} $$

The N-1 is important for when bias is a concern, which we will look at in detail in a seperate chapter, along with a more detailed coverage of variances and co-variances.

```python
import numpy
import matplotlib.pyplot as plt


# This is a probability density function for a gaussian. 
# We'll go over these in more details later.
def gaussian(x,mean,sigma):
    return numpy.exp(-((x-mean)**2)/(2*sigma**2))

x = numpy.arange(0,10,0.1)


## Plot three gaussians with some spread.


vgaussian = numpy.vectorize(gaussian)
gaussian1 = gaussian(x,5.0,1.0)
gaussian2 = gaussian(x,5.0,2.0)
gaussian3 = gaussian(x,5.0,5.0)


plt.plot(x,gaussian1,label='Sigma = 1.0')
plt.plot(x,gaussian2,label='Sigma = 2.0')
plt.plot(x,gaussian3,label='Sigma = 5.0')
plt.legend()
plt.show()
```
