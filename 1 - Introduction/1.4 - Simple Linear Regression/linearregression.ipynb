# Linear Regression

This is the first example of creating a model to fit to some data, and is amongst the simplest and most useful methods to use for many types of experimental analyses. We are trying to fit a straight line through our dataset:

$$ y = mx + c $$,

Where $m$ is the gradient of the function and $c$ is the y-intercept. Mathematically there are a few ways we can go about solving the dataset for the line. Mathematically the operation can be done by minimising a functional using euler lagrange:

$$ min \bigg[\Sigma^N_{i=0}\big(x - \bar{X}\big)^2 = \Sigma^N_{i=0}\big(y_{i} - mx_{i} - c\big)^2\bigg]$$

From this, and after some simplification we can derive $m$ and $c$ as:

$$ m = \frac{cov(X,Y)}{var(X,Y)} $$

$$ c = \bar{y} - m\bar{x} $$

It is worth pointing out that regression models can be more complicated than this, however this will be covered later.

```python
import numpy
import os
import matplotlib.pyplot as plt

# Again: numpy.cov does this nicely
def covariancef(X,Y,N):
    X_mean = numpy.mean(X)
    Y_mean = numpy.mean(Y)
    sum = 0.0
    for x,y in zip(X,Y):
        sum += (x - X_mean)*(y - Y_mean)
    return sum/N


covariance = [[1.0,0.9],
              [0.9,1.0]]
mean = [5.0,5.0]

samples = 500

samples = [numpy.random.multivariate_normal(mean,covariance) for i in range(samples)]
samples = numpy.array(samples)
   
m = covariancef(samples[:,0],samples[:,1],len(samples))#/covariance(samples[:,0],samples[:,0],len(samples))
c = numpy.mean(samples[:,1]) - m * numpy.mean(samples[:,0])

xvar = numpy.arange(0,10)

plt.plot(samples[:,0],samples[:,1],'x')
plt.plot(xvar,m*xvar + c)
plt.show() ## Show our points.


```
